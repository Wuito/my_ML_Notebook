```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

# Sigmoid

```python
#  Sigmoid 
def sigmoid(x):
    return 1 / (1 + tf.exp(-x))

#  Sigmoid 
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

#  x 
x_values = np.linspace(-7, 7, 200)
#  Sigmoid  x_values 
y_sigmoid = sigmoid(x_values)
#  Sigmoid  x_values 
y_derivative = sigmoid_derivative(x_values)
#  Sigmoid 
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(x_values, y_sigmoid, label='Sigmoid Function')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.legend()
#  Sigmoid 
plt.subplot(1, 2, 2)
plt.plot(x_values, y_derivative, label='Sigmoid Derivative')
plt.title('Sigmoid Derivative')
plt.xlabel('x')
plt.ylabel('sigmoid\'(x)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

# Tanh

```python
#  tanh 
def tanh(x):
    return tf.math.tanh(x)

#  tanh 
def tanh_derivative(x):
    return 1 - tf.math.square(tf.math.tanh(x))

#  x 
x_values = np.linspace(-7, 7, 200)

#  tanh  x_values 
y_tanh = tanh(x_values)

#  tanh  x_values 
y_derivative = tanh_derivative(x_values)

#  tanh 
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(x_values, y_tanh, label='tanh Function')
plt.title('tanh Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.legend()

#  tanh 
plt.subplot(1, 2, 2)
plt.plot(x_values, y_derivative, label='tanh Derivative')
plt.title('tanh Derivative')
plt.xlabel('x')
plt.ylabel('tanh\'(x)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

# ReLu

```python
#  ReLU 
def relu(x):
    return tf.nn.relu(x)

#  ReLU 
def relu_derivative(x):
    return tf.where(x > 0, 1.0, 0.0)

#  x 
x_values = np.linspace(-3, 3, 200)

#  ReLU  x_values 
y_relu = relu(x_values)

#  ReLU  x_values 
y_derivative = relu_derivative(x_values)

#  ReLU 
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(x_values, y_relu, label='ReLU Function')
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.legend()

#  ReLU 
plt.subplot(1, 2, 2)
plt.plot(x_values, y_derivative, label='ReLU Derivative')
plt.title('ReLU Derivative')
plt.xlabel('x')
plt.ylabel('ReLU\'(x)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

# LeakyRelu

```python
#  Leaky ReLU 
def leaky_relu(x, alpha=0.15):
    return tf.nn.leaky_relu(x, alpha)

#  Leaky ReLU 
def leaky_relu_derivative(x, alpha=0.15):
    return tf.where(x > 0, 1.0, alpha)

#  x 
x_values = np.linspace(-3, 3, 200)

#  Leaky ReLU  x_values 
y_leaky_relu = leaky_relu(x_values)

#  Leaky ReLU  x_values 
y_derivative = leaky_relu_derivative(x_values)

#  Leaky ReLU 
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(x_values, y_leaky_relu, label='Leaky ReLU Function')
plt.title('Leaky ReLU Function')
plt.xlabel('x')
plt.ylabel('Leaky ReLU(x)')
plt.grid(True)
plt.legend()

#  Leaky ReLU 
plt.subplot(1, 2, 2)
plt.plot(x_values, y_derivative, label='Leaky ReLU Derivative')
plt.title('Leaky ReLU Derivative')
plt.xlabel('x')
plt.ylabel('Leaky ReLU\'(x)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
